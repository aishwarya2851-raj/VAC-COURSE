```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))
plt.suptitle("final report") # Corrected from subtitle to suptitle

# Subplot 1: Scatter plot
plt.subplot(3,3,1)
plt.scatter(df['Time'], df['V22'], color='red')
plt.title("credit card fraud detection")
plt.xlabel("Time")
plt.ylabel("V22")

# Subplot 2: Pie chart
plt.subplot(3,3,2) # Changed to a new position
plt.pie(df['Time'])
plt.title("Time Distribution") # Added a title for clarity

# Subplot 3: Line plot
plt.subplot(3,3,3)
plt.plot(df["Time"])
plt.title("Time Series Plot") # Added a title for clarity

# Subplot 4: Histogram
plt.subplot(3,3,4)
plt.hist(df["V22"])
plt.title("V22 Histogram") # Added a title for clarity
plt.xticks([]) # Corrected from false to []

plt.tight_layout() # Adjust layout to prevent overlap
plt.savefig("credit card fraud detection.png")
plt.show()
```
